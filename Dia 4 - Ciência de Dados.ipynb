# **Conclusão do Dia 3:**

**Cluster 0:** São as turmas menos disputadas (Ociosas ou de Baixa Procura)


**Cluster 1:** São as turmas "padrão" da faculdade, em que a procura e as vagas são similares (Padrão da Universidade)

**Cluster 2:** São turmas com pouca oferta e muita demanda (Pós-Graduação ou Especializadas)

**Cluster 3:** São turmas com MUITA procura e poquíssimas vagas ofertadas (Super-Concorridas)

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt

df = pd.read_csv('turmas_limpo.csv')

features = ['capacidade_aluno', 'total_solicitacoes', 'ch_dedicada_periodo']
df_cluster = df[features].copy().dropna()
scaler = StandardScaler()
dados_escalonados = scaler.fit_transform(df_cluster)

# Aplicando K-Means com k=4
kmeans = KMeans(n_clusters=4, n_init=10, random_state=42)
df_cluster['cluster'] = kmeans.fit_predict(dados_escalonados)
```

# **Hipótese 1:** A Carga Horária do "Cluster de Pós-Graduação" (Cluster 2) é de fato superior?

Nós supomos que o Cluster 2 é de "Pós-Graduação" por causa de sua carga horária alta. Vamos testar se essa carga horária é estatisticamente maior do que a do Cluster das Turmas Padrão

**Hipótese Nula:** A carga horária média do Cluster 2 é menor ou igual à do Cluster 1.

**Hipótese Alternativa:** A carga horária média do Cluster 2 é maior que a do Cluster 1.

```python
# Prepara as amostras dos clusters
pos_grad_ch = df_cluster[df_cluster['cluster'] == 2]['ch_dedicada_periodo']
padrao_ch = df_cluster[df_cluster['cluster'] == 1]['ch_dedicada_periodo']

# Realiza o Teste t
stat, p_valor_bicaudal = ttest_ind(pos_grad_ch, padrao_ch, equal_var=False)
p_valor_unicaudal = p_valor_bicaudal / 2

if p_valor_unicaudal < 0.05 and pos_grad_ch.mean() > padrao_ch.mean():
    print("Conclusão: Hipótese confirmada! A carga horária do Cluster 2 é estatisticamente superior à do Cluster 1.")
else:
    print("Conclusão: Hipótese não confirmada. Não há evidências de que a carga horária do Cluster 2 seja superior.")
```

# **Hipótese 2:** As Turmas Super-Concorridas (Cluster 3) são significativamente diferentes em capacidade das Turmas Ociosas (Cluter 0)?

O Cluster 3 é extremamente procurado enquanto o Cluster 0 é prouco procurado. A capacidade da turma para esses dois grupos é diferente?

**Hipótese Nula:** A capacidade média do Cluster 3 é igual à do Cluster 0

**Hipótese Alternativa:** A capacidade média do Cluster 3 é diferente do Cluster 0

```python
# Prepara as amostras
super_concorridas_cap = df_cluster[df_cluster['cluster'] == 3]['capacidade_aluno']
ociosas_cap = df_cluster[df_cluster['cluster'] == 0]['capacidade_aluno']

# Realiza o Teste t
stat, p_valor = ttest_ind(super_concorridas_cap, ociosas_cap, equal_var=False)

if p_valor < 0.05:
    print("Conclusão: Hipótese confirmada! Existe uma diferença significativa na capacidade média entre as Turmas Super-Concorridas e as Ociosas.")
else:
    print("Conclusão: Hipótese não confirmada. Não há diferença estatística na capacidade média entre os dois grupos.")
```

# **Hipótese 3:** A baixa demanda as Turmas Ociosas (Cluster 0) é estatisticamente comprovada?

Vamos testar se o número de solicitações para estre grupo é menor do que a do Cluster Padrão (Cluster 1).

**Hipótese Nula:** A demanda média do Cluster 0 é maior ou igual à do Cluster 1.

**Hipótese Alternativa:** A demanda média do Cluster é menor do que a do Cluster 1.

```python
# Prepara as amostras
ociosas_demanda = df_cluster[df_cluster['cluster'] == 0]['total_solicitacoes']
padrao_demanda = df_cluster[df_cluster['cluster'] == 1]['total_solicitacoes']

# Realiza o Teste t
stat, p_valor_bicaudal = ttest_ind(ociosas_demanda, padrao_demanda, equal_var=False)
p_valor_unicaudal = p_valor_bicaudal / 2

if p_valor_unicaudal < 0.05 and ociosas_demanda.mean() < padrao_demanda.mean():
    print("Conclusão: Hipótese confirmada! A demanda pelo Cluster 0 é estatisticamente menor que a do Cluster 1, validando o perfil de 'ocioso'.")
else:
    print("Conclusão: Hipótese não confirmada. Não há evidências de que a demanda pelo Cluster 0 seja menor.")
```
