# Passo **1**

Primeiramente, iremos concatenar os arquivos

```python
import pandas as pd
import numpy as np

arquivos = [
    'turmas-2023.1.csv',
    'turmas-2023.2.csv',
    'turmas-2024-1.csv',
    'turmas-2024-2.csv'
]

turmas_dfs = []

for arquivo in arquivos:
  df_temp = pd.read_csv(arquivo, sep=';')
  turmas_dfs.append(df_temp)

df = pd.concat(turmas_dfs, ignore_index = True)
```

Dando uma olhadas inicial nos dados


```python
print("\nVisualizando a base com o display:\n")
display(df)
```

# **Passo 2**

Agora, iremos utilizar as funções que aprendemos em aula

```python
# info():
  # Mostra informações gerais do dataset como quantidade de colunas, quantidade de linhas, número de valores faltantes, tipo das colunas, etc.
print("Saída da função info():\n")
print(df.info())

# describe():
  # Mostra informações estatísticas do dataset, como média geral das colunas, desvio padrão geral das colunas, etc
print("\nSaída da função describe():\n")
display(df.describe())

# shape():
  # Mostra o formato do dataset: (linhas,colunas)
print("\nSaída da função shape():\n")
print(df.shape)

# função head()
  # Mostra as primeiras n linhas de um DataFrame (por padrão, 5)
print("Primeiras 3 linhas:\n")
display(df.head(3))

# função tail()
  # Mostra as últimas n linhas de um DataFrame (por padrão, 5)
print("\nÚltimas 3 linhas:\n")
display(df.tail(3))
```

```python
# Criar uma cópia para manter o DataFrame original intacto
df_limpo = df.copy()
```

# **Passo 3**

Como vimos anteriormente, há algumas colunas vazias

```python
# Contar valores nulos em cada coluna
print(df.isnull().sum())
```

**1.**
Colunas como: "convenio", "observacao" apresentam a maioria esmagadora das linhas vazias.  "id_turma_agrupadora" apresenta um problema similar, porém ela também é redundante, pois já existe uma coluna "agrupadora", que fornece a mesma informação de maneira mais clara e direta.

**2.**
"siape" e "matrícula_docente_externo" são mutualmente exclusivos.
Com isso iremos criar uma coluna "id_docente", que irá agrupar os valores não nulos dessas duas colunas.

**3.**
Embora "campus_turma", "local" e "descricao_horario" tenham muitos nulos, não podemos simplesmente remover essas colunas. Com isso, iremos preencher essas partes vazias com "NÃO INFORMADO".

**4.**
"total_solicitacoes" e "qtd_aulas_lancadas", quando há um valor nulo, podemos substituir por 0.
"capacidade_aluno", como há poucos valores ausentes, vamos utilizar a mediana.

```python
# 1. Remover colunas
df_limpo = df_limpo.drop(columns=['convenio', 'observacao', 'id_turma_agrupadora'])

# 2. Unificar colunas de docente
df_limpo['id_docente'] = df_limpo['siape'].fillna(df_limpo['matricula_docente_externo'])
df_limpo = df_limpo.drop(columns=['siape', 'matricula_docente_externo'])

# 3. Preencher colunas categóricas
colunas_categoricas_nulas = ['campus_turma', 'local', 'descricao_horario']
df_limpo[colunas_categoricas_nulas] = df_limpo[colunas_categoricas_nulas].fillna('NÃO INFORMADO')

# 4. Preencher colunas numéricas
df_limpo['total_solicitacoes'] = df_limpo['total_solicitacoes'].fillna(0)
df_limpo['qtd_aulas_lancadas'] = df_limpo['qtd_aulas_lancadas'].fillna(0)
mediana_capacidade = df_limpo['capacidade_aluno'].median()
df_limpo['capacidade_aluno'] = df_limpo['capacidade_aluno'].fillna(mediana_capacidade)

print("Passo 3: Tratamento de valores ausentes concluído.")
```

# **Passo 4**

De maneira geral, vamos apenas corrigir os valores para os tipos corretos

```python
# Converter colunas de data para o formato datetime
# errors='coerce' transforma entradas inválidas em NaT (Not a Time)
df_limpo['data_inicio'] = pd.to_datetime(df_limpo['data_inicio'], errors='coerce')
df_limpo['data_fim'] = pd.to_datetime(df_limpo['data_fim'], errors='coerce')
df_limpo['data_consolidacao'] = pd.to_datetime(df_limpo['data_consolidacao'], errors='coerce')

# Ajustar tipos de dados numéricos
# 'Int64' com 'I' maiúsculo permite a presença de valores nulos (NaT) no id_docente
df_limpo['id_docente'] = df_limpo['id_docente'].astype('Int64')
df_limpo['capacidade_aluno'] = df_limpo['capacidade_aluno'].astype(int)
df_limpo['total_solicitacoes'] = df_limpo['total_solicitacoes'].astype(int)
df_limpo['qtd_aulas_lancadas'] = df_limpo['qtd_aulas_lancadas'].astype(int)


print("Passo 4: Ajuste de tipos de dados concluído.")
df_limpo.info()
```

Conseguimos perceber que está dando tudo certo, por enquanto

# **Passo 5**

Verificar linhas duplicadas e removê-las

```python
# Verificar a quantidade de linhas duplicadas
duplicatas = df_limpo.duplicated().sum()
print(f"Número de linhas duplicadas encontradas: {duplicatas}")

# Remover as linhas duplicadas, se houver
if duplicatas > 0:
    df_limpo = df_limpo.drop_duplicates()
    print("Linhas duplicadas foram removidas.")

print(f"\nPasso 5: Remoção de duplicados concluído. Novo formato do DataFrame: {df_limpo.shape}")
```

# **Passo 6**

Vamos padronizar as colunas textuais, convertendo todas as letras para minúsculas e removendo espaços em branco no início e no fim

```python
# Selecionar colunas do tipo 'object' (texto)
colunas_texto = df_limpo.select_dtypes(include=['object']).columns

# Aplicar padronização
for coluna in colunas_texto:
    df_limpo[coluna] = df_limpo[coluna].str.lower().str.strip()

print("Passo 6: Padronização de texto concluído.")
display(df_limpo[['nivel_ensino', 'campus_turma', 'situacao_turma']].head())
```

# **Passo 7**


Vamos analisar os outlier utilizando o Boxplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o estilo dos gráficos
sns.set(style="whitegrid")

# Colunas para análise de outliers
colunas_para_boxplot = ['ch_dedicada_periodo', 'capacidade_aluno', 'total_solicitacoes', 'qtd_aulas_lancadas']

# Criar os boxplots
plt.figure(figsize=(15, 10))
for i, coluna in enumerate(colunas_para_boxplot, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=df_limpo[coluna])
    plt.title(f'Boxplot de {coluna}')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```
